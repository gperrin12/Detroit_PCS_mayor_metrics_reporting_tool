```python
import pandas as pd

pd.set_option('display.max_rows', 300)
pd.set_option('display.max_columns', 300)
```

```python
df = pd.read_csv('deduped_all_years.csv')
df.head()
```

```python
# Get counts by year
df['Created'] = pd.to_datetime(df['Created'])
df['year'] = df['Created'].dt.year
df['year'].value_counts()
```

```python
df['Sheet Name'].value_counts()
```

```python
sheet_names_cm8 = df['Sheet Name'].unique()
cm8_sheets = [name for name in sheet_names_cm8 if 'CM 8' in str(name)]
print("Sheet names containing 'CM 8':")
for name in cm8_sheets:
    print(name)
```

```python
cm8_df = df[df['Sheet Name'].isin(cm8_sheets)]
print("Value counts for 'Case Status' in CM 8 sheets:")
print(cm8_df['Case Status'].value_counts())
```

```python
cm9_sheets = [name for name in df['Sheet Name'].unique() if 'CM 9' in str(name)]
print("Sheet names containing 'CM 9':")
for name in cm9_sheets:
    print(name)
```
